# 2. Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên

```python
df[(df['type_of_land']=='Bán nhà riêng')][(df['ward_name']=='Phường Trung Liệt')|(df['ward_name']=='Phường Khâm Thiên')]
```

# 3. Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.

```python
df[['address','price','house_direction', 'balcony_direction']][(df['land_certificate'] == 'Sổ đỏ') & (df['bedroom']>3)]
```

# 4. Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.

```python
df.pivot_table(values='price', index='type_of_land', aggfunc={'sum','mean'})
```

# 5. Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường.

```python
df.pivot_table(values=['toilet','bedroom', 'floor'], index='ward_name', aggfunc=np.mean)
```

```python
df.groupby('ward_name')['bedroom', 'floor','toilet'].mean()
```
